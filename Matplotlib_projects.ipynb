Given File:
amazon_jobs_dataset.csv
It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. 


Problem Statement :
Plot the line graph between no. of Job postings with respect to year.
Print the year and the number of job posting as integer value.
Note: Year should be in ascending order.
Output Format :
year1 job_posting1
year2 job_posting2

```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
df=pd.read_csv('amazon_jobs_dataset.csv')
with open('amazon_jobs_dataset.csv','r',encoding ='ISO-8859-1') as file_obj:
    file_data=csv.DictReader(file_obj)
    year=[]
    for row in file_data:
        row['Posting_date']=row['Posting_date'].replace(' ','')
        year.append(row['Posting_date'].split(',')[1])
    np_year=np.array(year)
    values,counts=np.unique(np_year,return_counts=True)
    a=0
    for i in values:
        for j in counts[a:]:
            print(i,j)
            a=a+1
            break
    plt.plot(values,counts)
    plt.xlabel('Year')
    plt.ylabel('Number of job opening')
    plt.title('Year Vs Number of Job Posting')
    plt.show()
```

```python
Problem Statement :
Plot the Bar graph between Month vs Job Openings.
Print the month name and the number of job posting as integer value.
Order of months doesn't matter.
Output Format :
month_name1 job_posting1
month_name2 job_posting2
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
df=pd.read_csv('amazon_jobs_dataset.csv')
with open('amazon_jobs_dataset.csv','r',encoding ='ISO-8859-1') as file_obj:
    file_data=csv.DictReader(file_obj)
    month=[]
    for row in file_data:
        month.append(row['Posting_date'].split()[0])
    np_month=np.array(month)
    df['Month']=np_month
    count=df['Month'].value_counts()
    x=count.index
    y=count.values
    plt.bar(x,y)
    plt.xticks(rotation=40)
    plt.show()
    a=0
    for i in x:
        for j in y[a:]:
            print(i,j)
            a=a+1
            break
```

Problem Statement :
Plot the Pie chart between Indian cities vs No. of jobs opening.
Print the Indian cities and %age of Job distribution in India upto 2 decimal places.
Note: %age of Job distribution should be in descending order.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
df=pd.read_csv('amazon_jobs_dataset.csv')
with open('amazon_jobs_dataset.csv','r',encoding ='ISO-8859-1') as file_obj:
    file_data=csv.DictReader(file_obj)
    city=[]
    for row in file_data:
        if row['location'].split(',')[0]=='IN':
            ans=row['location'].split(',')[2].strip()
            city.append(ans)
    np_city=np.array(city)
    value,counts=np.unique(np_city,return_counts=True)
    d={}
    a=0
    for i in value:
        d[i]=counts[a]
        a=a+1
    value=list(value)
    counts=list(counts)
    counts.sort(reverse=True)
    new_value=[]
    for i in counts:
        for j in value:
            if d[j]==i:
                new_value.append(j)
                ans=(i/sum(counts))*100
                print(j,format(ans,".2f"))
                break
    plt.pie(counts,labels=new_value,autopct="%.2f%%")
    plt.show()
```

Problem Statement :
Plot the scatter graph between year vs No. of jobs opening related to Java.
Print the year and number of Jobs opening in Java Profile.
Note: Use the Keyword 'Java' or 'java' in Basic Qualification feature for finding the job opening related to Java Profile. Print the year in ascending order.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
df=pd.read_csv('amazon_jobs_dataset.csv')
with open('amazon_jobs_dataset.csv','r',encoding ='ISO-8859-1') as file_obj:
    file_data=csv.DictReader(file_obj)
    year=[]
    for row in file_data:
        if ('Java'in row['BASIC QUALIFICATIONS']) or ('java'in row['BASIC QUALIFICATIONS']):
            y=row['Posting_date'].split(',')[1]
            y=y.strip()
            year.append(y)
    np_year=np.array(year)
    values,counts=np.unique(np_year,return_counts=True)
    values=list(values)
    counts=list(counts)
    d={}
    a=0
    for i in values:
        d[i]=counts[a]
        a=a+1
#     print(d)
    values.sort()
    new_counts=[]
    for i in values:
        new_counts.append(d[i])
#     print(values)
#     print(new_counts)
    a=0
    for i in values:
        for j in new_counts[a:]:
            print(i,j)
            a=a+1
            break
plt.scatter(values,new_counts)
plt.xlabel('Year')
plt.ylabel('No.of jobs opening related to Java')
plt.show()
```
